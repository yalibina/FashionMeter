```python
# Base directory to save all styles
base_dir = "styles"
os.makedirs(base_dir, exist_ok=True)
```

```python
def get_json(links):
    failed_dict = {}
    # Iterate through the dictionary
    for key, url in links.items():
        # Extract the style (first word of the key)
        style = key.split()[0]
        
        # Create a directory for the style if it doesn't exist
        style_dir = os.path.join(base_dir, style)
        os.makedirs(style_dir, exist_ok=True)
        
        # Define the JSON output file path
        json_output = os.path.join(style_dir, f"{key.replace(' ', '_')}.json")
        
        # Scrape and save the JSON file
        try:
            PinterestDL.with_api(
                timeout=3,
                verbose=False,
            ).scrape_and_download(
                url=url,
                output_dir=style_dir,  # Save images to the style directory
                limit=200,  # Max number of images to download
                min_resolution=(224, 224),  # Minimum resolution for images
                json_output=json_output,  # Save JSON file with URLs
                dry_run=True,  # Set to False to download images
                add_captions=True,  # Add captions as metadata
            )
            print(f"Successfully scraped {url} and saved JSON to {json_output}")
        except Exception as e:
            print(f"Failed to scrape {key}: {e}")
            failed_dict[key] = url
    return failed_dict
```

```python
def calculate_md5(file_path: str) -> str:
    """
    Calculates the MD5 hash of a file.
    """
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()
```

```python
def read_json_files(base_dir: str) -> Dict[str, List[str]]:
    """
    Reads all JSON files in the base directory and extracts image URLs.
    Returns a dictionary where keys are directories and values are lists of image URLs.
    """
    image_urls = {}
    for root, _, files in os.walk(base_dir):
        for file in files:
            if file.endswith(".json"):
                json_path = os.path.join(root, file)
                with open(json_path, "r") as f:
                    try:
                        data = json.load(f)
                        # Extract image URLs from the JSON data
                        urls = [item["src"] for item in data if "src" in item]
                        # Use the relative path as the key to maintain directory structure
                        rel_path = os.path.relpath(root, base_dir)
                        if rel_path not in image_urls:
                            image_urls[rel_path] = []
                        image_urls[rel_path].extend(urls)
                    except json.JSONDecodeError as e:
                        print(f"Error reading {json_path}: {e}")
    return image_urls
```

```python
# Function to download images
def download_images(image_urls: Dict[str, List[str]], output_base_dir: str):
    """
    Downloads images from the provided URLs and saves them in the appropriate directories.
    Skips duplicate images across all classes using MD5 hashes.
    """
    downloaded_hashes: Set[str] = set()  # Track MD5 hashes of downloaded images

    for rel_path, urls in image_urls.items():
        # Create the output directory if it doesn't exist
        output_dir = os.path.join(output_base_dir, rel_path)
        os.makedirs(output_dir, exist_ok=True)

        for url in urls:
            # Extract the filename from the URL
            parsed_url = urlparse(url)
            filename = os.path.basename(parsed_url.path)
            file_path = os.path.join(output_dir, filename)

            # Download the image
            try:
                response = requests.get(url, timeout=10)
                response.raise_for_status()  # Raise an error for bad responses

                # Calculate the MD5 hash of the image content
                image_content = response.content
                image_hash = hashlib.md5(image_content).hexdigest()

                # Skip if the image has already been downloaded
                if image_hash in downloaded_hashes:
                    print(f"Skipping duplicate image: {filename} (hash: {image_hash})")
                    continue

                # Save the image
                with open(file_path, "wb") as f:
                    f.write(image_content)
                downloaded_hashes.add(image_hash)
                print(f"Downloaded {filename} to {file_path} (hash: {image_hash})")
            except requests.RequestException as e:
                print(f"Failed to download {url}: {e}")

```

```python
# Base directory containing JSON files
json_base_dir = "styles"
# Base directory to save downloaded images
output_base_dir = "downloaded_images"

# Step 1: Read JSON files and extract image URLs
image_urls = read_json_files(json_base_dir)

# Step 2: Download images
download_images(image_urls, output_base_dir)
```

```python
from collections import defaultdict

# Function to count images in each class
def count_images_per_class(base_dir: str) -> Dict[str, int]:
    """
    Counts the number of images in each class (subdirectory) within the base directory.
    Returns a dictionary where keys are class names and values are counts.
    """
    image_counts = defaultdict(int)  # Dictionary to store counts

    # Walk through the base directory
    for root, _, files in os.walk(base_dir):
        # Get the relative path (class name)
        rel_path = os.path.relpath(root, base_dir)
        if rel_path == ".":  # Skip the base directory itself
            continue

        # Count image files in the current directory
        image_count = sum(1 for file in files if file.lower().endswith((".jpg", ".jpeg", ".png", ".gif")))
        if image_count > 0:
            image_counts[rel_path] = image_count

    return image_counts


# Function to print the results
def print_image_counts(image_counts: Dict[str, int]):
    """
    Prints the number of images in each class in a readable format.
    """
    print("Number of images per class:")
    print("-" * 30)
    for class_name, count in image_counts.items():
        print(f"{class_name}: {count} images")
    print("-" * 30)
    total_images = sum(image_counts.values())
    print(f"Total images: {total_images}")


```

```python
 # Base directory containing downloaded images
base_dir = "downloaded_images"

# Step 1: Count images in each class
image_counts = count_images_per_class(base_dir)

# Step 2: Print the results
print_image_counts(image_counts)

```
